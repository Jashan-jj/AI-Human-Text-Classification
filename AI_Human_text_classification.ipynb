```python
import pandas as pd
import numpy as np
import re

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df= pd.read_csv('dataset/100K_Ai_Human.csv')
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'\W', ' ', text)   # remove special chars
    text = re.sub(r'\s+', ' ', text)  # remove extra spaces
    return text

df['clean_text'] = df['text'].apply(preprocess)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_text'], df['generated'],
    test_size=0.2,
    random_state=42
)
```

```python
vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2),   # unigram + bigram (important!)
    stop_words='english'
)

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
model = LogisticRegression()

model.fit(X_train_tfidf, y_train)
```

```python
y_pred = model.predict(X_test_tfidf)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
def predict_text(text):
    text = preprocess(text)
    vector = vectorizer.transform([text])
    prediction = model.predict(vector)[0]

    if prediction == 1:
        return "AI Generated"
    else:
        return "Human Written"

# Example
print(predict_text("Here’s why. I’m fully aware that most of what I read online these days is at the very least AI-assisted, if not entirely written by AI. Generally, I don’t mind. My main concern is whether what I’m reading is accurate or not. If I want to know how to operate my toaster, I don’t need to be assured that a human wrote the instructions. But when it comes to art and journalism, come on! With poetry, prose, etc., the whole point is the masterful placing of words in sentences, in paragraphs for the enjoyment and stirring of the reader."))
print(predict_text("We built a hybrid system combining TF-IDF, stylometric features, and embeddings to detect AI-generated text. The model analyzes linguistic patterns like sentence variability, vocabulary richness, and predictability to classify content."))
```
